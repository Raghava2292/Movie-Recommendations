```python
def convert_director(item):
    list1 = []
    for i in ast.literal_eval(item):
        if i['job'] == 'Director':
            list1.append(i['name'])
    return list1
```

```python
movies.crew = movies.crew.apply(convert_director)
movies
```

```python
movies.overview = movies.overview.apply(lambda x: x.split())
```

```python
movies
```

```python
movies.genres = movies.genres.apply(lambda x: [i.replace(" ", "") for i in x])
movies.keywords = movies.keywords.apply(lambda x: [i.replace(" ", "") for i in x])
movies.cast = movies.cast.apply(lambda x: [i.replace(" ", "") for i in x])
movies.crew = movies.crew.apply(lambda x: [i.replace(" ", "") for i in x])
```

```python
movies
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
movies_2 = movies[['movie_id', 'title', 'tags']]
movies_2
```

## Text preprocessing

```python
movies_2['tags'] = movies_2['tags'].apply(lambda x: " ".join(x))
movies_2
```

```python
movies_2['tags'] = movies_2['tags'].apply(lambda x: x.lower())
movies_2
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    list2 = []
    
    for i in text.split():
        list2.append(ps.stem(i))
    return " ".join(list2)
```

```python
movies_2['tags'] = movies_2['tags'].apply(stem)
movies_2
```

## Convert the tags into vectors using the vectorizer tool

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000, stop_words='english')
```

```python
vectors = cv.fit_transform(movies_2['tags']).toarray()
vectors
```

```python
cv.get_feature_names_out()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
sim_scores = cosine_similarity(vectors)
```

```python
sim_scores
```

```python
def recommend_movies(movie):
    movie_index = movies_2[movies_2['title'] == movie].index[0]
    distances = sim_scores[movie_index]
    movies_recommended = sorted(list(enumerate(distances)), reverse=True, key=lambda x:x[1])[1:6]
    for i in movies_recommended:
        print(movies_2['title'][i[0]])
```

```python
recommend_movies('The Dark Knight')
```
